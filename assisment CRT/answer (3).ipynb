```python
df=pd.read_csv('StudentData.csv')
```

```python
#23. Using NumPy, calculate the mean score of each subject.
mean_scores = df[['Math', 'Science', 'English', 'History']].mean()
mean_scores
```

```python
# 24.  Create a Pandas DataFrame for this dataset and calculate the average score of each student.
df['Average Score'] = df[['Math', 'Science', 'English', 'History']].mean(axis=1)
df[['Student Name', 'Average Score']]

```

```python
# 25.  Find the student with the highest overall score.
df['Total Score'] = df[['Math', 'Science', 'English', 'History']].sum(axis=1)

# Find the student with the highest overall score
top_student = df.loc[df['Total Score'].idxmax()]
top_student
```

```python
#26. Plot a scatter plot using Matplotlib to visualize Math vs Science scores.

plt.scatter(df['Math'], df['Science'], color='blue', marker='o')
plt.title('Math vs Science Scores')
plt.xlabel('Math Scores')
plt.ylabel('Science Scores')
plt.grid(True)
plt.show()
```

```python
# 27. Standardize the scores using NumPy by subtracting the mean and dividing by the standard deviation
standardized_scores = df[['Math', 'Science', 'English', 'History']].apply(
    lambda x: (x - np.mean(x)) / np.std(x)
)

# Add standardized scores to the DataFrame
df[['Math (Standardized)', 'Science (Standardized)', 'English (Standardized)', 'History (Standardized)']] = standardized_scores
df
```

```python
# 28. Visualize the total scores using a bar plot.
plt.bar(df['Student Name'], df['Total Score'], color='purple')
plt.title('Total Scores of Students')
plt.xlabel('Student Name')
plt.ylabel('Total Score')
plt.show()
```

```python
# 29. Generate a pie chart representing the subject-wise average scores.
mean_scores.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'], title='Subject-wise Average Scores')
plt.ylabel('')  # Remove the y-axis label for better visualization
plt.show()
```
